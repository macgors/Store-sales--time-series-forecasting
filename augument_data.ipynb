```python
import pandas as pd
```

```python
train = pd.read_csv('../project/train.csv', index_col='id')
test = pd.read_csv('../project/test.csv', index_col='id')
stores = pd.read_csv('../project/stores.csv')
holidays = pd.read_csv('../project/holidays_events.csv')
oil = pd.read_csv('../project/oil.csv')
```

```python
joined_train = train.join(oil, rsuffix='_oil').join(stores, rsuffix='_store').join(holidays, rsuffix='_holidays')

joined_train
```

```python
def is_holiday(row):
    if row['locale'] == 'National':
        return 1
    if row['locale'] == 'Local' and row['locale_name'] == row['city']:
        return 1
    if row['locale'] == 'Regional' and row['locale_name'] == row['state']:
        return 1

    return 0

joined_train['is_holiday'] = joined_train.apply(is_holiday, axis=1)

joined_train
```
